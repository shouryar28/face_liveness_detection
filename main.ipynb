```python
import cv2
import os

# Directory paths
input_dir = r'C:\Users\91790\Desktop\BioID-FaceDatabase-V1.2'
output_dir = r'C:\Users\91790\Desktop\BioID-FaceDatabase-V1.2_converted'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Function to convert .pgm to .jpg
def convert_pgm_to_jpg(input_path, output_path):
    img = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
    if img is not None:
        cv2.imwrite(output_path, img)
    else:
        print(f"Failed to read {input_path}")

# Convert and move images to appropriate subdirectories
for file_name in os.listdir(input_dir):
    if file_name.endswith('.pgm'):
        class_name = 'class_1'  # Update this if you have multiple classes or a naming convention
        class_dir = os.path.join(output_dir, class_name)
        os.makedirs(class_dir, exist_ok=True)
        img_path = os.path.join(input_dir, file_name)
        output_path = os.path.join(class_dir, file_name.replace('.pgm', '.jpg'))
        convert_pgm_to_jpg(img_path, output_path)

print("Conversion and organization completed.")
```

```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import EarlyStopping

# Define paths
train_data_path = 'C:/Users/91790/Desktop/BioID-FaceDatabase-V1.2_converted'

# Create data generators
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

# Check data generators
print("Train Generator Class Mode:", train_generator.class_mode)
print("Validation Generator Class Mode:", validation_generator.class_mode)

# Define the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Print model summary
model.summary()

# Train the model
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10,  # Adjust based on your needs
    callbacks=[early_stopping]
)

# Save the model
model.save('face_liveness_model.h5')

# Evaluate the model
loss, accuracy = model.evaluate(validation_generator)
print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")
```
